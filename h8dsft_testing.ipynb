```python
class Buku:
    def __init__(self, judul, pengarang, isbn):
        """Function ini bertugas untuk menampung judul, pengarang dan ISBN (sebagai unique identifier).
        -------------------------------------------------------------------------------------------------------
        judul : Judul buku
        pengarang : Pengarang buku
        """
        # Menampung value Judul, Pengarang dan ISBN
        self.judul = judul
        self.pengarang = pengarang
        self.isbn = isbn

class Perpus:
    def __init__(self):
        self.katalog = []

    def tambah_buku(self, buku: Buku):
        """Function ini bertugas untuk menambahkan buku ke dalam perpustakaan.
        """
        self.katalog.append(buku)

    def cari_buku(self, judul):
        """Function ini bertugas untuk mencari buku berdasarkan judul.
        ----------------------------------------------------------------------
        judul : Judul buku
        """
        found = []
        for buku in self.katalog:
            if buku.judul == judul:
                found.append(buku)
        return found
    
    def cari_buku_pengarang(self, pengarang):
        """Function ini bertugas untuk mencari buku berdasarkan pengarang.
        ----------------------------------------------------------------------
        pengarang : Pengarang buku
        """
        found = []
        for buku in self.katalog:
            if buku.pengarang == pengarang:
                found.append(buku)
        return found

    def remove_buku(self, isbn):
        """Function ini bertugas untuk menghapus buku dari daftar perpustakaan.
        ----------------------------------------------------------------------
        isbn : ISBN buku
        """
        count = 0
        for buku in self.katalog:
            if buku.isbn == isbn:
                self.katalog.pop(count)
            else:
                count += 1
                continue

    def display_buku(self):
        """Function ini bertugas untuk menampilkan buku yang ada di daftar perpustakaan.
        --------------------------------------------------------------------------------
        """
        for buku in self.katalog:
            print(f"ISBN Buku: {buku.isbn}")
            print(f"{buku.judul} oleh {buku.pengarang}")

    def menu(self):
        while True:
            print("1. Tambah buku")
            print("2. Cari buku")
            print("3. Hapus Buku")
            print("4. Tampilkan buku")
            print("5. Exit")

            choice = input("Masukkan pilihan: ")

            if choice == '1':
                x = 1
                while x == 1:
                    judul = input("Masukkan judul: ")
                    pengarang = input("Masukkan nama pengarang: ")
                    isbn = input("Masukkan ISBN: ")
                    # Melakukan Exception handling untuk menjawab poin ketiga dalam NGC
                    try:
                        cekangka = int(isbn)
                    except ValueError:
                        print("ISBN harus angka!")
                        continue
                    else:
                        buku = Buku(judul, pengarang, isbn)
                        self.tambah_buku(buku)
                        print("Buku berhasil ditambah!")
                        x = 0

            elif choice == '2':
                pilihan = input('Berdasarkan judul atau pengarang : ')
                if pilihan == 'pengarang':
                    input_pengarang = input("Masukkan nama pengarang : ")
                    result = self.cari_buku_pengarang(pengarang=input_pengarang)
                elif pilihan == 'judul':
                    input_judul = input("Masukkan judul: ")
                    result = self.cari_buku(judul=input_judul)

                for i in result:
                    print(f"{i.judul} oleh {i.pengarang} (ISBN : {i.isbn})")

            elif choice == '3':
                input_isbn = input("Masukkan ISBN buku : ")
                self.remove_buku(input_isbn)

            elif choice == '4':
                self.display_buku()

            elif choice == '5':
                break
```

```python
import unittest

class TestLibrary(unittest.TestCase):

    def test_tambah_buku(self):
        # Input value
        perpus = Perpus()
        buku = Buku('The Great Gatsby', 'F. Scott Fitzgerald', 2)
        perpus.tambah_buku(buku)
        # Cek hasil
        self.assertEqual(len(perpus.katalog), 1)

        self.assertIn(buku, perpus.katalog)

    def test_cari_buku(self):
        # Input value
        perpus = Perpus()
        buku = Buku('a', 'a', 1)
        buku2 = Buku('a', 'b', 2)
        buku3 = Buku('b', 'a', 3)
        for b in [buku, buku2, buku3]:
            perpus.tambah_buku(b)
        result = perpus.cari_buku(judul='a')
        result2 = perpus.cari_buku_pengarang(pengarang='b')
        # Cek hasil
        self.assertEqual(len(result), 2)
        self.assertIn(buku2, result2)

    def test_hapus_buku(self):
        # Input value
        perpus = Perpus()
        buku = Buku('a', 'a', 1)
        buku2 = Buku('a', 'b', 2)
        buku3 = Buku('b', 'a', 3)
        for b in [buku, buku2, buku3]:
            perpus.tambah_buku(b)
        perpus.remove_buku(isbn=3)
        # Cek hasil
        self.assertEqual(len(perpus.katalog), 2)

unittest.main(argv=[''], exit=False)
```
