# Task

You will use `austin_bikeshare` dataset from Google BigQuery Public Dataset.

- BigQuery initiation

```python
from google.colab import auth
from google.cloud import bigquery

import pandas as pd
auth.authenticate_user()
project_id = "crafty-dynamics-260602"
client = bigquery.Client(project=project_id)
```

- How many tables are in the `austin_bikeshare` dataset?

```python
# Query
# Reference https://kontext.tech/article/676/get-list-of-tablesviews-and-schema-in-bigquery
query = "SELECT COUNT(*) AS jumlah_tabel FROM `bigquery-public-data.austin_bikeshare.__TABLES__`"

# Run query
query_job = client.query(query).to_dataframe()
query_job
```

- How many columns in the `bikeshare_stations` table have `STRING` data?

```python
# Query
# Reference https://stackoverflow.com/questions/51262224/bigquery-possible-to-query-for-column-datatype
query = "SELECT count(*) as jumlah_datatype FROM bigquery-public-data.austin_bikeshare.INFORMATION_SCHEMA.COLUMNS WHERE table_name = 'bikeshare_stations' and data_type = 'STRING'"

# Run query
query_job = client.query(query).to_dataframe()
query_job
```

- Retrieve all trips where the `bike_id` is not null and the duration is greater than 1000 minutes.

```python
# Query
query = "SELECT * FROM bigquery-public-data.austin_bikeshare.bikeshare_trips WHERE bike_id IS NOT NULL AND duration_minutes > 1000"

# Run query
query_job = client.query(query).to_dataframe()
query_job
```

- Filter trips where the `subscriber_type` is "Walk Up" and the `end_station_name` have "Street" or "St." in their name.

```python
query = "SELECT * FROM bigquery-public-data.austin_bikeshare.bikeshare_trips WHERE subscriber_type = 'Walk Up' AND end_station_name like '%Street' OR end_station_name like '%St.'"

# Run query
query_job = client.query(query).to_dataframe()
query_job
```
